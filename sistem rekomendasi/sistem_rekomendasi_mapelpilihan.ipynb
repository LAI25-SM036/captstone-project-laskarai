```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import joblib
```

```python
# 1. Prepare Item Unique Flags (3 packages x 9 flags)
item_flags = item_df.groupby('package_name').agg({
    'math_adv': 'max',
    'physics': 'max',
    'chemistry': 'max',
    'biology': 'max',
    'history_adv': 'max',
    'economics': 'max',
    'sociology': 'max',
    'geography': 'max',
    'informatics': 'max'
}).reindex(['IPA', 'IPS', 'Campuran']).reset_index()
```

```python
# Konversi ke matrix (3 x 9)
X_item_flags = item_flags[['math_adv', 'physics', 'chemistry', 'biology',
                           'history_adv', 'economics', 'sociology', 'geography', 'informatics']].values

# 2. Prepare User Interest Vectors (350 x 9) dari user_df
user_interest = user_df[['interest_math', 'interest_physics', 'interest_chemistry',
                         'interest_biology', 'interest_history', 'interest_economics',
                         'interest_sociology', 'interest_geography', 'interest_informatics']].values
```

```python
# Standardisasi interest agar sebanding dengan flags biner
scaler_interest = StandardScaler()
X_user_interest = scaler_interest.fit_transform(user_interest)
```

```python
# Simpan scaler_interest untuk keperluan inference nanti
joblib.dump(scaler_interest, 'scaler_interest.pkl')

# 3. Hitung Cosine Similarity (350 users x 3 packages)
sim_matrix = cosine_similarity(X_user_interest, X_item_flags)

# 4. Prediksi paket: pilih indeks tertinggi dan +1 untuk label {1,2,3}
y_pred = np.argmax(sim_matrix, axis=1) + 1
```

```python
y_true = y_df['selected_label'].values
print("=== Classification Report ===")
print(classification_report(y_true, y_pred, zero_division=0))
print("=== Confusion Matrix ===")
print(confusion_matrix(y_true, y_pred))
print("=== Overall Accuracy ===")
print(accuracy_score(y_true, y_pred))
```

```python
scaler_interest = joblib.load('scaler_interest.pkl')
item_flags = pd.read_pickle('item_flags.pkl')  # DataFrame 3 baris

# Siapkan item flag matrix dan mapping label  nama paket
X_item_flags = item_flags[['math_adv', 'physics', 'chemistry', 'biology',
                           'history_adv', 'economics', 'sociology', 'geography', 'informatics']].values
label_to_package = {1: 'IPA', 2: 'IPS', 3: 'Campuran'}

def predict_from_excel(excel_path):
    # Baca file Excel
    df_new = pd.read_excel(excel_path)

    # Kolom interest yang dibutuhkan
    required_interest_cols = [
        'interest_math', 'interest_physics', 'interest_chemistry',
        'interest_biology', 'interest_history', 'interest_economics',
        'interest_sociology', 'interest_geography', 'interest_informatics'
    ]
    missing = set(required_interest_cols) - set(df_new.columns)
    if missing:
        raise ValueError(f"Missing interest columns: {missing}")

    # Extract dan scale interest
    user_new_interest = df_new[required_interest_cols].values
    X_new_interest = scaler_interest.transform(user_new_interest)

    # Hitung cosine similarity dengan item flags
    sim_new = cosine_similarity(X_new_interest, X_item_flags)  # shape (n_new, 3)

    # Prediksi label = argmax + 1
    y_new_pred = np.argmax(sim_new, axis=1) + 1
    df_new['predicted_label'] = y_new_pred
    df_new['predicted_package'] = df_new['predicted_label'].map(label_to_package)

    return df_new[['student_id', 'predicted_label', 'predicted_package']]
```
